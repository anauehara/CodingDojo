```python
#resposta
import turtle


pigHead = turtle.Turtle()
pigHead.pendown()
pigHead.left(90)
for i in range(6):
    for j in range(3):
        pigHead.forward(100)
        pigHead.right(120)
    pigHead.left(60)
    
pigHead.penup()
pigHead.hideturtle()
turtle.done()
```
